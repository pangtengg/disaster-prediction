```python
import pandas as pd
from pycaret.regression import *
```

```python
df = pd.read_csv('global_disaster.csv')

# Feature engineering
df['month'] = pd.to_datetime(df['date']).dt.month
df['year'] = pd.to_datetime(df['date']).dt.year
df.drop(columns=['date'], inplace=True)
```

```python
# Setup AutoML - predicting response_time_hours
s = setup(df, target='response_time_hours', 
          categorical_features=['country', 'disaster_type'],
          session_id=42)

# Compare all models automatically
best_model = compare_models()

# Tune the best one
tuned = tune_model(best_model)

# Save model
save_model(tuned, 'disaster_response_model')
# This creates: disaster_response_model.pkl
```
